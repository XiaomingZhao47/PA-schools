```python
import pandas as pd
import sqlite3

excel_file_path = '/Users/jellyzhao/Downloads/data.xlsx'

df = pd.read_excel(excel_file_path)

print("Original columns:", df.columns)

df_pivoted = df.pivot_table(index=['DistrictName', 'Name', 'AUN', 'Schl'], 
                            columns='DataElement', 
                            values='DisplayValue', 
                            aggfunc='first').reset_index()

df_pivoted.columns = [col.replace(' ', '_').replace('(', '').replace(')', '') for col in df_pivoted.columns]

print("Pivoted DataFrame columns:", df_pivoted.columns)

conn = sqlite3.connect('/Users/jellyzhao/schools_facts.db')
c = conn.cursor()

c.execute('''
    CREATE TABLE IF NOT EXISTS SchoolFacts (
        School_Name TEXT PRIMARY KEY,
        School_Address_Street TEXT,
        School_Address_City TEXT,
        School_Address_State TEXT,
        School_Zip_Code TEXT,
        Website TEXT,
        Telephone_Number TEXT,
        Grades_Offered TEXT,
        Title_I_School TEXT,
        School_Enrollment INTEGER,
        Percent_of_Gifted_Students REAL,
        Intermediate_Unit_Name TEXT,
        Intermediate_Unit_Website TEXT,
        American_Indian_Alaskan_Native REAL,
        Asian REAL,
        Native_Hawaiian_or_other_Pacific_Islander REAL,
        Black_African_American REAL,
        Hispanic REAL,
        White REAL,
        Two_or_More_Races REAL,
        Economically_Disadvantaged REAL,
        English_Learner REAL,
        Special_Education REAL,
        Female_School REAL,
        Male_School REAL,
        Career_and_Technical_Programs TEXT
    )
''')

for index, row in df_pivoted.iterrows():
    print(f"Inserting row for school: {row.get('Name')}")
    
    c.execute('''
        INSERT OR REPLACE INTO SchoolFacts (
            School_Name, School_Address_Street, School_Address_City, School_Address_State, School_Zip_Code,
            Website, Telephone_Number, Grades_Offered, Title_I_School, School_Enrollment,
            Percent_of_Gifted_Students, Intermediate_Unit_Name, Intermediate_Unit_Website,
            American_Indian_Alaskan_Native, Asian, Native_Hawaiian_or_other_Pacific_Islander,
            Black_African_American, Hispanic, White, Two_or_More_Races,
            Economically_Disadvantaged, English_Learner, Special_Education, Female_School,
            Male_School, Career_and_Technical_Programs
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (
        row.get('Name', None), 
        row.get('School_Address_Street', None), 
        row.get('School_Address_City', None),
        row.get('School_Address_State', None), 
        row.get('School_Zip_Code', None), 
        row.get('Website', None),
        row.get('Telephone_Number', None), 
        row.get('Grades_Offered', None), 
        row.get('Title_I_School', None),
        row.get('School_Enrollment', None), 
        row.get('Percent_of_Gifted_Students', None),
        row.get('Intermediate_Unit_Name', None), 
        row.get('Intermediate_Unit_Website', None),
        row.get('American_Indian/Alaskan_Native_', None), 
        row.get('Asian_', None),
        row.get('Native_Hawaiian_or_other_Pacific_Islander_', None), 
        row.get('Black/African_American_', None),
        row.get('Hispanic_', None), 
        row.get('White_', None), 
        row.get('2_or_More_Races', None),
        row.get('Economically_Disadvantaged', None), 
        row.get('English_Learner', None),
        row.get('Special_Education', None), 
        row.get('Female_School', None),
        row.get('Male_School', None), 
        row.get('Career_and_Technical_Programs', None)
    ))

conn.commit()
conn.close()
```
